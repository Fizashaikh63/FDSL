```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Hospital patient dataset
data = {
    'Patient_Name': ['Amit', 'Priya', 'Rohan', 'Sneha', 'Vikram', 'Anjali', 'Raj', 'Neha'],
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],  # Qualitative
    'Department': ['Cardiology', 'Orthopedics', 'Neurology', 'Pediatrics', 'Cardiology', 'Orthopedics', 'Neurology', 'Pediatrics'],  # Qualitative
    'Age': [45, 38, 60, 12, 50, 42, 55, 8],  # Quantitative
    'Bill_Amount': [55000, 48000, 72000, 30000, 60000, 45000, 70000, 25000],  # Quantitative
    'Days_Admitted': [5, 4, 7, 3, 6, 4, 8, 2]  # Quantitative
}

df = pd.DataFrame(data)
print(df)



```

```python
# --- QUALITATIVE ANALYSIS ---
print("Qualitative Analysis:")
print(df['Gender'].value_counts())      # Frequency of categories
print(df['Department'].value_counts())  # Frequency of departments

# Bar chart for Gender distribution
df['Gender'].value_counts().plot(kind='bar', color=['pink', 'lightblue'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

```

```python
# --- QUANTITATIVE ANALYSIS ---
print("\nQuantitative Analysis:")
print(df[['Age', 'Bill_Amount', 'Days_Admitted']].describe())  # Summary statistics

# Histogram for Age
df['Age'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Scatter plot: Age vs Bill Amount
plt.scatter(df['Age'], df['Bill_Amount'], color='green')
plt.title("Age vs Bill Amount")
plt.xlabel("Age")
plt.ylabel("Bill Amount")
plt.show()

# Scatter plot: Days Admitted vs Bill Amount
plt.scatter(df['Days_Admitted'], df['Bill_Amount'], color='orange')
plt.title("Days Admitted vs Bill Amount")
plt.xlabel("Days Admitted")
plt.ylabel("Bill Amount")
plt.show()
```
